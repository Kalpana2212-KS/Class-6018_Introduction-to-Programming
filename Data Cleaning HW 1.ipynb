```python
# Import libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# check if data is imported accurately
flights_data = pd.read_csv('flights.csv')
flights_data.head(5)
```

# Questions



```python
# Question 1: How many flights were there from JFK to SLC? int

# data filtered by comparing values from columns 'origin' and 'dest' in dataframe
q_1 = flights_data[(flights_data['origin'] == 'JFK') & (flights_data['dest'] == 'SLC')] 

#count the number of rows that match with the criteria
filtered_data1 = len(q_1)

# print result
print(f"The number of flights from JFK to SLC is: {filtered_data1}")
```

```python
# Question 2: How many airlines fly to SLC? int

# data filtered for, destination to SLC and column 'carrier' which is airline identifier
q_2 = flights_data[flights_data['dest'] == 'SLC']['carrier'].unique()

# count the number of unique carriers
filtered_data2 = len(q_2)

# print the results
print(f"The number of unique airlines that fly to SLC are: {filtered_data2}")
```

```python
# Question 3: What is the average arrival delay for flights to RDU? float

# filter rows where destination is RDU
# extract arr_delay column which indicates delay in arrival
# calculate the average arrival delay

q_3 = flights_data[flights_data['dest'] == 'RDU']['arr_delay'].mean()
print(f"The average arrival delay for flights to RDU is: {q_3}")
```

```python
# Question 4: What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  float

# filter flights to SEA
sea_flights = flights_data[flights_data['dest'] == 'SEA']

nyc_to_sea = sea_flights[sea_flights['origin'].isin(['LGA', 'JFK'])].shape[0] #gives the number of rows with LGA and JFK

#calculate proportion
q_4 = nyc_to_sea/sea_flights.shape[0] * 100
print(q_4)
```

```python
# For question 5, first, create a 'date' column.

flights_data['date'] = flights_data['year'].astype(str) + '/' + flights_data['month'].astype(str) + '/' + flights_data['day'].astype(str)

# check first few rows if 'date' column in updated.
flights_data.head(5)
```

```python
# Question 5: Which date has the largest average depature delay? 
# Find unique dates
unique_dates = flights_data['date'].unique()

# Dictionary to hold the average delays for each date
average_delays = {}

# Calculate the mean delay for each date
for date in unique_dates:
    # Filter the rows for the current date
    daily_flights = flights_data[flights_data['date'] == date]
    
    # Calculate the mean delay for these rows
    mean_delay = daily_flights['dep_delay'].mean()
    
    # Store the result
    average_delays[date] = mean_delay

# Now find the date with the maximum average delay
max_delay_date = max(average_delays, key=average_delays.get)
max_delay_value = average_delays[max_delay_date]

print(f"The date with the largest average departure delay is {max_delay_date} with an average delay of {max_delay_value}.")
```

```python
# Alternate method using group.by
# group the date column and find the mean dep delay for each date
grouped = flights_data.groupby('date')['dep_delay'].mean()

# find the date with largest dep_delay
max_delay_date = grouped.idxmax()
max_delay_value = grouped.max()

print(f"The date with largest average departure delay is {max_delay_date} with an average delay of {max_delay_value}.")
```

```python
# Question 6: Which date has the largest average arrival delay?
# Find unique dates
unique_dates = flights_data['date'].unique()

# Dictionary to hold the average delays for each date
average_arr_delays = {}

# Calculate the mean delay for each date
for date in unique_dates:
    # Filter the rows for the current date
    daily_flights = flights_data[flights_data['date'] == date]
    
    # Calculate the mean delay for these rows
    mean_arr_delay = daily_flights['arr_delay'].mean()
    
    # Store the result
    average_arr_delays[date] = mean_arr_delay

# Now find the date with the maximum average arrival delay
max_arr_delay_date = max(average_arr_delays, key=average_arr_delays.get)
print(f"The date with the largest average arrival delay is {max_arr_delay_date}.")
```

```python
# Alternate method: using groupby
avg_delays = flights_data.groupby('date')['arr_delay'].mean()
q_6 = avg_delays.idxmax()
print(q_6)
```

```python
# For question 7, create a new column 'speed'
flights_data['speed'] = flights_data['distance']/ flights_data['air_time']

# check if 'speed' column is added 
flights_data.head(5)
```

```python
# Question 7:  Which flight departing LGA or JFK in 2013 flew the fastest?

# filter dataset for flights originating only from 'LGA' and 'JFK'
# Then, filter for date 2013.
filtered_flights_data = flights_data[(flights_data['origin'].isin(['LGA', 'JFK'])) & (flights_data['date'].str.contains('2013'))]

# From,the filtered data, sort vales bases on speed
q_7 = filtered_flights_data.sort_values(by = 'speed', ascending = False).iloc[0]
print(q_7)
```

```python
# Question 9: How many observations were made in Feburary? Int
weather_data_np = weather_data_pd.to_numpy()
# For column 3, which is 'month', and february represented as 2.0
feb_obs = weather_data_np[weather_data_np[:, 3] == 2.0]

# count the number of observations made in feb
num_feb_obs = len(feb_obs)
print(f"There are {num_feb_obs} observations made in the month February")
```

```python
# Question 10: What was the mean for humidity in February? Float

# Extract 'humid' column values for feb rows
humid_values_feb = feb_obs[:, 8]

# calculate mean humidity for feb
q_10 = np.mean(humid_values_feb)
print(f"The mean humidity in February is: {q_10}")
```

```python
#Question 11: What was the std for humidity in February? Float

# find standard deviation for humidity values in feb
q_11 = np.std(humid_values_feb)
print(f"The standard deviation of humidity in february is: {q_11}")
```
